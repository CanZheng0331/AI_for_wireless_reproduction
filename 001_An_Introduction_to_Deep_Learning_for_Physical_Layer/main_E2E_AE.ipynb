# Autoencoders for end-to-end communications systems
Pytorch implementation

```python
from figures import (
    TrainConfig,
    get_device,
    plot_figure_3,
    plot_figure_4,
    run_figure_3a,
    run_figure_3b,
    run_figure_4,
    set_seed,
)

set_seed(42)
device = get_device()
print(f"Using device: {device}")
```

```python
# Hyper-parameters
train_cfg_default = TrainConfig(
    epochs=80,
    batch_size=512,
    iterations_per_epoch=200,
    eval_iterations=50,
    train_ebno_db=7.0,
    lr=0.001,
    test_num_samples=1_000_000,
    test_batch_size=50_000,
)

train_cfg_88 = TrainConfig(
    epochs=120,
    batch_size=512,
    iterations_per_epoch=200,
    eval_iterations=50,
    train_ebno_db=7.0,
    lr=0.001,
    test_num_samples=1_000_000,
    test_batch_size=50_000,
)
```

```python
# Figure 3(a) entry
fig3a = run_figure_3a(
    device=device,
    train_cfg=train_cfg_default,
    snr_start=-4,
    snr_end=9,
    snr_step=1,
    mld_num_samples=1_000_000,
    mld_batch_size=50_000,
)
```

```python
# Figure 3(b) entry
fig3b = run_figure_3b(
    device=device,
    train_cfg_default=train_cfg_default,
    train_cfg_88=train_cfg_88,
    snr_start=-2,
    snr_end=11,
    snr_step=1,
)
```

```python
# Plot both subfigures
plot_figure_3(fig3a, fig3b)
```

```python
# Figure 4 entry
fig4 = run_figure_4(
    device=device,
    train_cfg_default=train_cfg_default,
    train_cfg_fig4=TrainConfig(
        epochs=150,
        batch_size=512,
        iterations_per_epoch=100,
        eval_iterations=50,
        train_ebno_db=7.0,
        lr=0.005,
        test_num_samples=1_000_000,
        test_batch_size=50_000,
    ),
    tsne_perplexity=30,
    tsne_random_state=42,
    tsne_samples_per_msg=60,
    tsne_snr_db=7.0,
)
```

```python
# Plot Figure 4
plot_figure_4(fig4)
```
